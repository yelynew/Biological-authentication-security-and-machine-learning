```

from skimage.morphology import binary_opening, binary_closing,square

import matplotlib.image as mpimg

import matplotlib.pylab as plt

from skimage.morphology import skeletonize

import numpy as np

import math


cn1_1 = 0

cn1_3 = 0

d = 0
# Mth와 Dth 설정
mth  = 50

dth = 70

frr = 0

# 1(1),1(2)~8(2)까지 불러온다.
for m in range(1,9):

  for n in range(1,3):

    dcnt = 0

    mlist1_1 = []

    image_name1 = f'/gdrive/My Drive/ColabNote/your address/{m} ({n}).bmp'

    img1_1 = mpimg.imread(image_name1)

    img_bin1_1 = np.zeros((144,144))

    img_bin1_1[img1_1<128] = 1

    img_bin1_1[img1_1>=128] = 0
# bth= 127 기준으로 이진화
    for i in range(144):

      for j in range(144):

        if((i<25 or i>120) or (j<25 or j>120)):
# 엣지를 잘라준다. (0~24, 120~144 범위의 숫자를 0으로 변경)
          img_bin1_1[i][j] = 0

       

   
# 열림을 진행한 후 닫힘 진행
    p1_1 = binary_closing(binary_opening(img_bin1_1))

          

    

     
# skeletonize를 활용한 세선화
    p1_1 = skeletonize(p1_1)
# 3*3 구현 
    for i in range(1, 143):

        for j in range(1, 143):

          cn1_1 = (abs((int)(p1_1[i,j+1])-(int)(p1_1[i-1,j+1]))+abs((int)(p1_1[i-1,j+1])-(int)(p1_1[i-1,j]))+abs((int)(p1_1[i-1,j])-(int)(p1_1[i-1,j-1]))+abs((int)(p1_1[i-1,j-1])-(int)(p1_1[i,j-1]))+abs((int)(p1_1[i,j-1])-(int)(p1_1[i+1,j-1]))+abs((int)(p1_1[i+1,j-1])-(int)(p1_1[i+1,j]))+abs((int)(p1_1[i+1,j])-(int)(p1_1[i+1,j+1]))+abs((int)(p1_1[i+1,j+1])-(int)(p1_1[i,j+1]))) / 2

          if cn1_1 == 3:

            mlist1_1.append([i,j,cn1_1])
# 현재 자신의 번호의 (3)~(10)의 번호를 받아온다.
    for k in range(3,11):

      mcnt = 0

      mlist1_3 = []

      image_name2 = f'/gdrive/My Drive/ColabNote/your address/{m} ({k}).bmp'

      img1_3 = mpimg.imread(image_name2)

      img_bin1_3 = np.zeros((144,144))
# Bth=127 기준 이진화
      img_bin1_3[img1_3<128] = 1

      img_bin1_3[img1_3>=128] = 0

    

            
# 엣지를 잘라준다. (0~24, 120~144 범위의 숫자를 0으로 변경)
      for i in range(144):

        for j in range(144):

          if((i<25 or i>120) or (j<25 or j>120)):

            img_bin1_3[i][j] = 0

     

    

# 열림을 진행한 후 닫힘 진행
      p1_3 = binary_closing(binary_opening(img_bin1_3))

      
# skeletonize를 활용한 세선화
      p1_3 = skeletonize(p1_3)

      

    

      


         
# 3*3 구현
      for i in range(1,143):

        for j in range(1,143):

          cn1_3 = (abs((int)(p1_3[i,j+1])-(int)(p1_3[i-1,j+1]))+abs((int)(p1_3[i-1,j+1])-(int)(p1_3[i-1,j]))+abs((int)(p1_3[i-1,j])-(int)(p1_3[i-1,j-1]))+abs((int)(p1_3[i-1,j-1])-(int)(p1_3[i,j-1]))+abs((int)(p1_3[i,j-1])-(int)(p1_3[i+1,j-1]))+abs((int)(p1_3[i+1,j-1])-(int)(p1_3[i+1,j]))+abs((int)(p1_3[i+1,j])-(int)(p1_3[i+1,j+1]))+abs((int)(p1_3[i+1,j+1])-(int)(p1_3[i,j+1]))) / 2

          
#cn=3일때 특이점이라고 취급하고 mlist에 넣어준다.
          if cn1_3 == 3:

            mlist1_3.append([i,j,cn1_3])  

      

      

      

      

      #print(len(mlist1_1))

      #print(len(mlist1_3))

     

      
#특이점으로 추출된 좌표들 사이의 거리를 구한다.
      for l in range(len(mlist1_1)):

        for s in range(len(mlist1_3)):

          d = math.sqrt((mlist1_1[l][0]-mlist1_3[s][0]) ** 2 + (mlist1_1[l][1] - mlist1_3[s][1]) ** 2)
#임계치보다 거리가 작을때 mcnt 1 추가 (매칭) 
          if d <= mth:

            mcnt += 1
#실제로 매칭된 경우의 수 / 조합으로 만들 수 있는 모든 매칭의 경우의 수 * 100 -> 매칭률

      mr = mcnt/(len(mlist1_1)*len(mlist1_3)) * 100


      

#임계치보다 매칭률이 작을때가 본인인데 거부되는 경우 이므로 count
      if mr < dth:

        dcnt += 1
#샘플에 따른 매칭 확률 출력 
    print(dcnt/8)

    frr +=  dcnt / 8

    #print(frr)       
# 퍼센테지로 전체 오거부 확률 출력
frr = frr / (8 * 2) *100
print("전체 오거부율")
print(frr)
```
